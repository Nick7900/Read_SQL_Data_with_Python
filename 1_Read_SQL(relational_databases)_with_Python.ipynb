# **1. How to read data before Pandas (old procedure)**

## **Read data from SQL database**
Reading data from SQL relational databases is relatively simple, and pandas supports a number of methods for doing so.

We'll start with SQLite because it's a built-in Python package and we don't need anything else installed.

````python
import sqlite3
import pandas as pd
````

In order to work with a SQLite database from Python, we first have to connect to it. We can do that using the connect function, which returns a Connection object.

````python
conn = sqlite3.connect('chinook.db')
````

Once we have a ```Connection object```, we can then create a ```Cursor``` object. Cursors allow us to execute SQL queries against a database:MM

````python
cur = conn.cursor()
````

The ```Cursor``` created has a method ```execute```, which will receive SQL parameters to run against the database.

The code below will fetch the first 5 rows from the employees table:

````python
cur.execute('SELECT * FROM employees LIMIT 5;')
````

As you can see, we did not assign the result of the preceding query to a variable. This is because we need to run another command to get the results.

We can use the ```fetchall``` method to get all of the results of a query:

````python
results = cur.fetchall()
results
````

The results are returned as a list of tuples. Each tuple corresponds to a database row that we accessed. Dealing with data in this manner is torturous.

We'd have to manually insert column headers and parse the data. Fortunately, the pandas library has a simpler method, which we'll look at in the following section.

````python

df = pd.DataFrame(results)
df
````


Before we proceed, it is a good practice to close any open ```Connection``` or ```Cursor``` objects. This prevents the SQLite database from being locked. When a SQLite database is locked, you may be unable to update it and may encounter errors. We can close the ```Cursor``` and the ```Connection``` by doing the following:



````python
cur.close()
conn.close()
````

# **2. Read SQL database with Pandas**

## **Using pandas ```read_sql``` method**
The pandas read ```sql function``` can be used to read the results of a SQL query directly into a pandas DataFrame. The code below will run the same query as before, but it will return a ```DataFrame```. It has several advantages over the previous query:

* It does not involve the creation of a ```Cursor``` object or the use of fetchall at the end.
* The names of the table's headers are automatically read in.
* It generates a ```DataFrame```, allowing us to quickly explore the data.a.

````python
conn = sqlite3.connect('chinook.db') # Connect to the database
df = pd.read_sql('SELECT * FROM employees;', conn) # Read the data from the SQL database
df.head()
````

If we want to get a little more fancy, we can specify the index column and convert the dates into Python-date objects rather than strings.

````python
df =pd.read_sql('SELECT * FROM employees;', conn,
                index_col ='EmployeeId',
                parse_dates =["BirthDate", "HireDate"])
df.head()
````

````python
df.info() # get information about the table
````

````python
# Call not a Number value
df['ReportsTo'].isna().sum()
````

````python
df['ReportsTo']
````

## **Using pandas ```read_sql_query``` method**
 The ```read_sql method``` is actually a wrapper around for two other methods:
 * ```read_sql_query```
 * ```read_sql_table```

We can get the same result using ```read_sql_query``` method, since it is the default behaviour:

````python
conn = sqlite3.connect('chinook.db') # Connect to SQL database
df = pd.read_sql_query('SELECT * FROM employees', conn) #Read database into pandas
df.head()
````

````python
df =pd.read_sql_query('SELECT * FROM employees;', conn,
                index_col ='EmployeeId',
                parse_dates =["BirthDate", "HireDate"])
df.head()
````

````python
df.info()
````

## **Using read_sql_table method**
In contrast the ```read_sql_table``` can read an entire table by just passing its name and it will automatically give all the information in the database.
The ```read_sql_table``` is a useful function, but it works only with [SQLAlchemy](https://www.sqlalchemy.org/), a Python SQL Toolkit and Object Relational Mapper.

This is just a demonstration of its usage where we read the whole employees table.

````python
from sqlalchemy import create_engine
````

````python
engine = create_engine('sqlite:///chinook.db') # create engine object
conn = engine.connect() # Connect to database
df = pd.read_sql_table('employees', con=conn,
                       index_col="EmployeeId",
                       parse_dates=["BirthDate", "HireDate"])
df.head()
````

````python
conn.close() # Close the connection
````
